```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy_utils import create_database, database_exists

import pymysql
pymysql.install_as_MySQLdb()
```

```python
engine = create_engine(connection)
```

```python
# Check if the database exists. If not, create it.
if database_exists(connection) == False:
  create_database(connection)
else:
  print('The database already exists')
```

```python
basics = pd.read_csv('title.basics.csv.gz')
```

```python
basics.head(5)
```

```python
## create a col with a list of genres
basics['genres_split'] = basics['genres'].str.split(',')
basics
```

```python
exploded_genres = basics.explode('genres_split')
exploded_genres

```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
print(unique_genres)
```

```python
title_genres =exploded_genres[['tconst', 'genres_split']].copy()
title_genres.head(5)
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
basics['genre_id'] = 0
#

```

```python
genres =basics[['genres', 'genre_id']].copy()
genres.head(5)
```

```python
title_genres= basics[['tconst', 'genre_id']].copy()
title_genres.head(5)
```

```python
basics.head(5)
```

```python
for i in basics['runtimeMinutes']:
    if i == 'Talk-Show':
        basics['runtimeMinutes']= 1
        break
```

```python
#basics['endYear'].replace("\\N" "", inplace=True)
#basics.dropna(subset=['endYear'], inplace=True)
basics.info()
```

```python
## Example
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'runtimeMinutes':Integer()}

```

```python
# Save to sql with dtype and index=False
basics.to_sql('title_basics',engine,dtype=df_schema,if_exists='replace',index=False)
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```
