```python
df_merged['has_financial_info'] = df_merged[['budget', 'revenue']]>0
num_movies_with_financial_info = df_merged['has_financial_info'].sum()
print(f'There are {num_movies_with_financial_info} movies with at least some valid financial information.')
```

```python
ax = sns.countplot(data=df_merged, x='has_financial_info')
```

```python
certification_counts = df_merged['certification'].value_counts()

print(certification_counts)
```

```python
ax = sns.countplot(data=df_merged, x='certification')
```

```python
average_revenue = df_merged.groupby('certification')['revenue'].mean()

print(average_revenue)
```

```python
# Plot the average of each as a bar
fig, ax = plt.subplots()
ax.bar(average_revenue.index, average_revenue.values)
ax.set_ylabel('revenue')
ax.set_xlabel('certification')
```

```python
average_budget = df_merged.groupby('certification')['budget'].mean()

print(average_budget)
```

```python
# Plot the average of each as a bar
fig, ax = plt.subplots()
ax.bar(average_budget.index, average_budget.values)
ax.set_ylabel('budget')
ax.set_xlabel('certification')
```
